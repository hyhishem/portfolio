## <h3 id="section3.1">  4.1. Matrice de corrélation entre l’ensemble des indicateurs  </h3>   

 <p> Dans cette partie nous allons calculer des matrice de corrélation afin de selectioner que des indicateur peu corrélé. Pour cela nous allons utliser la methode $\textcolor{red}{.corr(method=)}$ en specifiant la méthode, "spearman" ou "pearson"
 
 
 </p>

## <h4 id="section4.1.1">  4.1.1  Methode de pearson </h3>   

 <p>
 La methode pearson est adapté au corelation linéaire pour des grandeurs quantitatives. Elle ne permet pas de detecter des corrélations non linéaire. Elle reste néamoins sensible au outliers.
 </p>

```python
df_Data_Country_Indicator_pearson=df_Data_Country_Indicator.iloc[:,1:].corr(method="pearson").abs()
plt.figure(figsize=(6, 4))
sns.heatmap(
    data=df_Data_Country_Indicator_pearson,    # tableau ou DataFrame à afficher
    cmap="coolwarm",     # palette de couleurs
    vmin=0, vmax=1,     # bornes min/max de l’échelle des couleurs
    linewidths=0.5,      # épaisseur des lignes de séparation
    cbar_kws={"shrink": 0.8, "label": "Corrélation"}  # options de la barre
)
plt.title("Matrice de corrélation pearson", fontsize=11)
plt.show()
```

## <h4 id="section4.1.1">  4.1.2  Methode de spearman </h3>   

 <p>
  La methode spearman est adapté aux corelations linéaire ou non mais forcement monotone. Etant basée sur le rang,  elle est adaptée aux grandeurs quantitatives ou qualitative. Elle ne permet pas de detecter des corrélations non monotonne.
     La methode de spearman est mons sensible aux outliers que la methode pearson.
 </p>

```python
df_Data_Country_Indicator_spearman=df_Data_Country_Indicator.iloc[:,1:].corr(method="spearman").abs()
plt.figure(figsize=(6, 4))
sns.heatmap(
    data=df_Data_Country_Indicator_spearman,   
    cmap="coolwarm",     # palette de couleurs
    vmin=0, vmax=1,     # bornes min/max de l’échelle des couleurs
    linewidths=0.5,      # épaisseur des lignes de séparation
    cbar_kws={"shrink": 0.8, "label": "Corrélation"}  # options de la barre
)
plt.title("Matrice de corrélation spearman", fontsize=11)
plt.show()


```

 ## <h4 id="section4.2.3"> 4.2.3 Méthode quantitative, sur la base des indicateurs: </h4>   

<p>
Pour decrire quantitativement ces données nous allons expliciter une fonction dépandant de ces quatres. On va normaliser les indicateurs afin de pouvoir ponderé pour donner d''avantage de poids à un indicateur en particulier.

LA fonction cout sera défini par:

$$ F_n()= \alpha \times X+(1+\frac{P_{grow} }{100})^n \times [ \beta \times Z \times Y + \gamma \times Z] $$

- Les variable  X (internet), Y (PIB) et Z (etudiants) sont normalisé par la valeure maximale de chaque indicateur.
- La variable P (évolution de la population) en pourcentage.
- $ \alpha, \beta \text{ et } \gamma $  sont les coeficients poids.
- Le parametre n nous permetera de tenir compte de l'évolution de la population dans les années futur.
    
- $\alpha=0.25$, $\beta=2$  et $\gamma=5$. On donne ainsi plus de poids à la population d'étudient et au PIB. 

   
</p>


```python
df_Data_Country_Indicator_norm=df_Data_Country_Indicator.copy()

df_Data_Country_Indicator_norm[['Internet', 'PIB', 'Population_Etudiant']]=df_Data_Country_Indicator[['Internet', 'PIB', 
                                        'Population_Etudiant']]/df_Data_Country_Indicator[['Internet',
                                        'PIB', 'Population_Etudiant']].max()
Alpha=0.25
Beta=2
Gamma=5
P=df_Data_Country_Indicator_norm['Population_Dynamique']
X=df_Data_Country_Indicator_norm['Internet']
Y=df_Data_Country_Indicator_norm['PIB']
Z=df_Data_Country_Indicator_norm['Population_Etudiant']
n=0

F_0=df_Data_Country_Indicator_norm.copy()
Countr_score=pd.DataFrame()

for n in range(0, 21, 5):
    F_0[f'F{n}']=Alpha*X+((1+P/100)**n)*Z*(Gamma + Beta*Y) 
    Countr_score[f'{2025+n}']=F_0[['Pays', f'F{n}']].sort_values(by=f'F{n}', ascending=False).iloc[0:10,0].reset_index(drop=True)
    
Countr_score
```
